```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
stem_port=PorterStemmer()
```

```python
import nltk
nltk.download('stopwords')
```

```python
def stemming(content):
  stemmed_content=re.sub('[^a-zA-Z]',' ',content)
  stemmed_content=stemmed_content.lower()
  stemmed_content=stemmed_content.split()
  stemmed_content=[stem_port.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content=' '.join(stemmed_content)
  return stemmed_content
```

```python
df['text']=df['text'].apply(stemming)
```

```python
df.head()
```

```python
x=df['text'].values
y=df['label'].values
```

```python
x
```

```python
y
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer= TfidfVectorizer()
vectorizer.fit(x)
x=vectorizer.transform(x)
```

```python
print(x)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
ypred= model.predict(x_test)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```

```python
text=df['message to examine']
text
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer

stemmer= SnowballStemmer('english')
tk= TweetTokenizer()
```

```python
text = text.apply(lambda x:[stemmer.stem(i.lower()) for i in tk.tokenize(x)]).apply(lambda y:' '.join(y))
text
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

sw= set(stopwords.words('english'))
```

```python
text= text.apply(lambda x:[i for i in tk.tokenize(x) if i not in sw]).apply(lambda x:' '.join(x))
```

```python
text.shape
```

```python
text
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer= TfidfVectorizer()
train_vec= vectorizer.fit_transform(text)
print(train_vec)
```

```python
train_vec.shape
```

```python
x= train_vec
y= df['label (depression result)'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
ypred= model.predict(x_test)
ypred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred)
```
