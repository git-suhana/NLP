```python
from tensorflow.keras.datasets import fashion_mnist
```

```python
import numpy as np
```

```python
data= fashion_mnist.load_data()
```

3 type image:

```python
#preprocessing:
(x_train,y_train),(x_test,y_test)=data
```

```python
y_train.shape
```

```python
x_test
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[1])  ##displays the second image in the training set
```

```python
from tensorflow.keras.models import Sequential   #IMPORT SEQUENTIAL MODEL
from tensorflow.keras.layers import Flatten, Dense     #IMPORTS FLATTEN AND DENSE LAYERS, ESSENTIAL BUILDING BLOCKS OF NEURAL NETWORKS

from tensorflow.keras.optimizers import Adam    #IMPORTS ADAM OPTIMIZER,USED TO UPDATE MODELS WEIGHT DURING TRAINING
from tensorflow.keras.losses import SparseCategoricalCrossentropy      #IMPORT THE LOSS FUNCTION USED TO MEASURE THE DIFFERENCE B/W THE MODELS PREDICTIOS AND OUTPUT
```

```python
#building a model

model= Sequential()    #creates a sequential model
model.add(Flatten(input_shape=(28,28,1)))   #adds a flatten layer to the model to convert 28x28 images into 1d array.
model.add(Dense(units=600, activation='relu'))    #adds a dense layer with neurons and relu activation function.
model.add(Dense(units=300, activation='relu'))   #adds a dense layer with neurons.
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=10, activation='sigmoid'))   #adds a dense layer with neurons and sigmoid activation function.
```

```python
model.compile(optimizer=Adam(), loss=SparseCategoricalCrossentropy(), metrics=['accuracy'])
#model.compile()- preparing model for training
#optimizer='adam': specifies the adam optimizer for updating the model's weights
#loss= SparseCategoricalCrossentropy(): specifies the loss function
#metrics=['accuracy']: specifies the metrics to track during training, accuracy
```

```python
#training the model

performance= model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
#epochs= iterations over entire dataset
#validation data= this validation set is used to evaluate the model's performanceduring the taining process
```

```python
performance.history
#attribute is a dictionary that stores the training and validation metrics recorded during the training purpose.
#value of the loss function calculated on the validation data during training.
```

```python
import matplotlib.pyplot as plt
plt.plot(performance.history['accuracy'],color='r', label='Training')
plt.plot(performance.history['val_accuracy'], color='g', label='Testing')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('ACCURACY CURVE')
```

```python
import tensorflow
from tensorflow.keras.datasets import cifar100
import numpy as np
import matplotlib.pyplot as plt
```

```python
data= cifar100.load_data()
data
```

```python
(X_train,y_train),(X_test,y_test)=data
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[1])
```

```python
from tensorflow.keras.models import Sequential   #IMPORT SEQUENTIAL MODEL
from tensorflow.keras.layers import Flatten, Dense     #IMPORTS FLATTEN AND DENSE LAYERS, ESSENTIAL BUILDING BLOCKS OF NEURAL NETWORKS

from tensorflow.keras.optimizers import Adam    #IMPORTS ADAM OPTIMIZER,USED TO UPDATE MODELS WEIGHT DURING TRAINING
from tensorflow.keras.losses import SparseCategoricalCrossentropy
```

```python
model= Sequential()    #creates a sequential model
model.add(Flatten(input_shape=(32,32,3)))   #adds a flatten layer to the model to convert 28x28 images into 1d array.
model.add(Dense(units=600, activation='relu'))    #adds a dense layer with neurons and relu activation function.
model.add(Dense(units=300, activation='relu'))   #adds a dense layer with neurons.
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=50, activation='relu'))
model.add(Dense(units=100, activation='sigmoid'))
```

```python
model.compile(optimizer=Adam(), loss=SparseCategoricalCrossentropy(), metrics=['accuracy'])
```

```python
performance= model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))
```

```python
performance.history
```

```python
import matplotlib.pyplot as plt
plt.plot(performance.history['accuracy'],color='r', label='Training')
plt.plot(performance.history['val_accuracy'], color='g', label='Testing')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('ACCURACY CURVE')
```
