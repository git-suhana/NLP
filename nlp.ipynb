```python
from nltk.util import ngrams
import nltk

sentence = 'this is my first sentence. A basic sentence'
tokens= nltk.word_tokenize(sentence)
tokens
```

```python
#pass the tokenized sentence as fist positional argument
Ngrams= ngrams(tokens,n=3)

for grams in Ngrams:
  print(grams)
```

stemming:

```python
from nltk.stem import SnowballStemmer
sb = SnowballStemmer('english')

words= ['eight','clock','program','programs','programing','programmers','thursday','morning','hello']
for w in words:
  print(w, ':', sb.stem(w))
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

words= ['eight','clock','program','programs','programing','programmers','thursday','morning','hello']
print ('word, Porter Stem')

for w in words:
  print(w, ':', ps.stem(w))
```

Lemmatization:

```python
import nltk
nltk.download('wordnet')

from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()

print('rocks', lem.lemmatize('rocks'))
print('thought', lem.lemmatize('thought'))
```

```python
#POS(Part of speech) Tagging

#In natural language processing, part of speech tagging is the process of assigning gramatical categories(like noun, verb, adjective, etc) to each word in a sentence. This information is crucial
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()

#lemmatize as nouns
print('rocks', lem.lemmatize('rocks'))
print('thought', lem.lemmatize('thought'))

#lemmatize as verbs

print('rocks as verb:', lem.lemmatize('rocks', pos='v'))
print('thought as verb:', lem.lemmatize('thought', pos='v'))
print('played as verb:', lem.lemmatize('played', pos='v'))
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()

#lemmatize as nouns, verbs, adjectives
print('rocks as noun', lem.lemmatize('rocks', pos='n'))
print('rocks as verb', lem.lemmatize('rocks', pos='v'))
print('rocks as adjective', lem.lemmatize('rocks', pos='a'))
print('better as adjective', lem.lemmatize('better', pos='a'))
print('better as adverb', lem.lemmatize('better', pos='r'))
```
